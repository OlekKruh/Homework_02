```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
file_adres = 'D:/MyProjects/GoIT_DS/Homework_02/bestsellers with categories.csv'
books_df = pd.read_csv(file_adres)

print(books_df.head())
```

```python
num_rows, num_columns = books_df.shape
print(f'Rows: {num_rows}')
print(f'Columns: {num_columns}')

unique_books = books_df['Name'].unique()
print(f'Unique books: {len(unique_books)}')
```

```python
# QUESTION: O ilu książkach przechowywane są dane w zbiorze danych?
# ANSWER: In the DataFrame, there are 351 unique book titles.
```

```python
print(f'Current columns names: \n{books_df.columns}')
books_df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
print(f'New columns names: \n{books_df.columns}')
```

```python
na_in_df = books_df.isna().sum()
print(f'NaN in columns:\n{na_in_df}\n')
nulls_in_df = books_df.isnull().sum()
print(f'Nulls in columns:\n{nulls_in_df}\n')
zero_in_df = books_df.apply(lambda x: (x == 0).sum())
print(f'Zero values in columns:\n{zero_in_df}')
```

```python
# QUESTION: Czy są jakieś zmienne z brakującymi wartościami? (Таk / Nie)
# ANSWER: There are no variables with missing values. Also, there are no cells with NaN or empty values cells in the books_df.
```

```python
unique_genre = books_df['genre'].unique()
print(f'Unique genre: {unique_genre}')
```

```python
# QUESTION: Jakie są unikalne gatunki?
# ANSWER: The 'genre' column has only two unique values, 'Non Fiction' & 'Fiction'.
```

```python
plt.figure(figsize=(20, 10))
books_df['price'].plot(kind='hist', bins=20, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Book Prices')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.gca().yaxis.set_major_locator(MultipleLocator(20))
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```

```python
price_min = books_df['price'].min()
price_max = books_df['price'].max()
price_mean = books_df['price'].mean()
price_median = books_df['price'].median()

print(f'Minimum price: {price_min}\nMaximum price: {price_max}\nMean price: {price_mean}\nMedian price: {price_median}')
```
