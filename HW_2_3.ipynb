```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import seaborn as sns
```

```python
file_adres = 'D:/MyProjects/GoIT_DS/Homework_02/bestsellers with categories.csv'
books_df = pd.read_csv(file_adres)

print(books_df.head())
```

```python
num_rows, num_columns = books_df.shape
print(f'Rows: {num_rows}')
print(f'Columns: {num_columns}')

unique_books = books_df['Name'].unique()
print(f'Unique books: {len(unique_books)}')
```

```python
# QUESTION: O ilu książkach przechowywane są dane w zbiorze danych?
# ANSWER: In the DataFrame, there are 351 unique book titles.
```

```python
print(f'Current columns names: \n{books_df.columns}')
books_df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
print(f'New columns names: \n{books_df.columns}')
```

```python
na_in_df = books_df.isna().sum()
print(f'NaN in columns:\n{na_in_df}\n')
nulls_in_df = books_df.isnull().sum()
print(f'Nulls in columns:\n{nulls_in_df}\n')
zero_in_df = books_df.apply(lambda x: (x == 0).sum())
print(f'Zero values in columns:\n{zero_in_df}')
```

```python
# QUESTION: Czy są jakieś zmienne z brakującymi wartościami? (Таk / Nie)
# ANSWER: There are no variables with missing values. Also, there are no cells with NaN or empty values cells in the books_df.
```

```python
unique_genre = books_df['genre'].unique()
print(f'Unique genre: {unique_genre}')
```

```python
# QUESTION: Jakie są unikalne gatunki?
# ANSWER: The 'genre' column has only two unique values, 'Non Fiction' & 'Fiction'.
```

```python
plt.figure(figsize=(20, 10))
books_df['price'].plot(kind='hist', bins=20, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Book Prices')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.gca().yaxis.set_major_locator(MultipleLocator(20))
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```

```python
price_min = books_df['price'].min()
price_max = books_df['price'].max()
price_mean = books_df['price'].mean()
price_median = books_df['price'].median()

print(f'Minimum price: {price_min}\nMaximum price: {price_max}\nMean price: {price_mean}\nMedian price: {price_median}')
```

```python
user_rating_max = books_df['user_rating'].max()

print(f'Maximum user rating: {user_rating_max}\n')
```

```python
fiction_books = books_df[books_df['genre'] == 'Fiction']
aggregated_fiction = fiction_books.groupby('genre')['price'].agg(max_price='max', min_price='min')
non_fiction_books = books_df[books_df['genre'] == 'Non Fiction']
aggregated_non_fiction = non_fiction_books.groupby('genre')['price'].agg(max_price='max', min_price='min')

print("Aggregated Fiction Prices:")
print(f'{aggregated_fiction}\n')
print("Aggregated Non Fiction Prices:")
print(f'{aggregated_non_fiction}')
```

```python
author_book_counts = books_df.groupby('author').agg(book_count=('name', 'count')).reset_index()
author_book_counts_sorted = author_book_counts.sort_values(by='book_count', ascending=False)

print(author_book_counts_sorted)
```

```python
# QUESTION: Jaki jest wymiar tabeli?
# ANSWER: 248 rows x 2 columns
# QUESTION: Który autor ma najwięcej książek?
# ANSWER: Jeff Kinney
# QUESTION: Ile jest książek tego autora?
# ANSWER: 12
```

```python
author_avg_rating = books_df.groupby('author').agg(avg_rating=('user_rating', 'mean')).reset_index()
author_avg_rating_sorted = author_avg_rating.sort_values(by='avg_rating', ascending=True)

print(author_avg_rating_sorted)
```

```python
# QUESTION: Który autor ma najniższą średnią ocenę?
# ANSWER: Donna Tartt
# QUESTION: Jaka jest średnia ocena tego autora?
# ANSWER: 3.9
```

```python
author_book_counts = books_df.groupby('author').agg(book_count=('name', 'count')).reset_index()

author_avg_rating = books_df.groupby('author').agg(avg_rating=('user_rating', 'mean')).reset_index()

combined_df = pd.concat([author_book_counts, author_avg_rating['avg_rating']], axis=1)

sorted_combined_df = combined_df.sort_values(by=['book_count', 'avg_rating'])

print(sorted_combined_df.head())
```

```python
# QUESTION: Który autor jest pierwszy na liście?
# ANSWER: Muriel Barbery
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='price', data=books_df)
plt.title('Box Plot of Book Prices by Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
average_rating_by_year = books_df.groupby('year')['user_rating'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='user_rating', data=average_rating_by_year, marker='o', color='green')
plt.title('Average User Rating by Year')
plt.xlabel('Year')
plt.ylabel('Average User Rating')
plt.grid(True)
plt.show()
```
